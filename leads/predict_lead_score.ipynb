## Um “lead” representa um cliente potencial interessado em comprar seu produto/serviço.

Normalmente, quando você adquire um lead, seja por meio de um serviço de terceiros ou ao executar uma campanha de marketing por conta própria, ele geralmente inclui informações como:

- Nome e detalhes de contato do lead
- Atributos do lead (demográficos, sociais, preferências do cliente)
- Fonte de origem (Anúncio no Facebook, página de destino do site, terceiros, etc.)
- Tempo gasto no site, número de cliques, etc.
- Detalhes de referência, etc.

## Lead Generation ou Geração do Lead

A geração de leads é o início do interesse ou da consulta do cliente sobre os produtos ou serviços do seu negócio. Os leads são criados com o objetivo de converter esse interesse ou consulta em vendas. Existe um número ilimitado de empresas terceirizadas na internet que prometem gerar os melhores leads. No entanto, você também pode fazer isso por conta própria, executando campanhas de marketing. Os métodos para gerar leads geralmente se enquadram na categoria de publicidade, mas também podem incluir fontes não pagas, como resultados orgânicos de motores de busca ou indicações de clientes existentes.

## Lead Qualification ou Ranqueamento do Lead

A qualificação de leads refere-se ao processo de determinar quais potenciais clientes têm maior probabilidade de realizar uma compra. É uma parte essencial do funil de vendas, que geralmente recebe muitos leads, mas converte apenas uma fração deles. Em termos simples, a qualificação de leads significa avaliar e priorizar os leads para determinar a probabilidade de conversão, permitindo que os departamentos de marketing e vendas foquem nos leads prioritários em vez de abordar todos os leads, que muitas vezes podem estar na casa dos milhares.

## Lead Conversion ou Conversão em Vendas

A conversão de leads é a fase em que você finalmente transforma um lead qualificado em um cliente pagante. Ela envolve todas as práticas de marketing que estimulam o desejo de comprar um produto ou serviço e incentivam o lead a tomar uma decisão de compra. Esta é a fase de monetização ou fechamento, e o resultado dessa etapa geralmente define o sucesso da campanha de marketing como um todo.

## O que significa ranquear um lead?

Imagine que você tenha um CRM com muitos leads cadastrados mas sua equipe de vendas não é suficiente para abordar todos estes leads no prazo necessário. Desta forma a empresa terá de priorizar os leads com maior probabilidade de conversão. Uma forma eficiente de se fazer isso **de forma automática** é utilizar **Machine Learning** para ranquear os leads de tal forma que a equipe de vendas terá maior probabilidade de sucesso.

```python
df = pd.read_csv('./../dados/leads-dataset/Leads.csv')
df.head()
```

```python
from collections import Counter
```

```python
## 1 indica conversão pelo time de vendas
## 0 indica que não houve conversão
Counter(df['Converted'])
```

```python
from pycaret.classification import ClassificationExperiment
exp = ClassificationExperiment()
```

```python
exp.setup(df, target = 'Converted', ignore_features = ['Prospect ID', 'Lead Number'],session_id = 100)
```

```python
melhor_modelo = exp.compare_models(sort = 'F1')
```

```python
exp.plot_model(melhor_modelo, plot = 'feature')
```

```python
exp.interpret_model(melhor_modelo)
```

Explica o impacto de cada feature (variável) no modelo preditivo. 

### 1. **Eixo Y (Lista de Features)**
Cada linha no eixo Y representa uma feature do modelo. Elas estão ordenadas pela importância, ou seja, as que estão no topo têm maior impacto na previsão.

### 2. **Eixo X (Valores SHAP)**
O eixo X mostra o impacto dos valores SHAP em uma previsão individual:
- **Valores positivos**: Aumentam a probabilidade do resultado ser positivo (ou a classe-alvo).
- **Valores negativos**: Reduzem a probabilidade do resultado ser positivo.

Quanto maior o valor absoluto no eixo X, maior é a influência da feature na previsão.

### 3. **Cores (Feature Value)**
- A cor representa o valor da feature.
  - **Azul**: Valores baixos da feature.
  - **Rosa/vermelho**: Valores altos da feature.

### 4. **Pontos no Gráfico**
Cada ponto representa uma observação (exemplo individual do conjunto de dados). A posição no eixo X indica o impacto do valor daquela feature na previsão para aquele exemplo.

---

### Interpretação de Exemplos do Gráfico
1. **Tags (Top Feature)**
   - Essa é a feature mais importante para o modelo.
   - Valores altos (vermelhos) aumentam o impacto positivo (para a direita), enquanto valores baixos (azuis) têm impacto negativo.

2. **Total Time Spent on Website**
   - Valores altos dessa variável têm um impacto positivo na previsão, sugerindo que passar mais tempo no site está associado a uma maior probabilidade de conversão.

3. **Lead Profile_Select**
   - Pode indicar que, dependendo do perfil do lead, o impacto varia, sendo valores altos associados a impactos mais positivos.

---

### Como Usar Essa Informação
1. **Otimizar a Estratégia**:
   - Focar nas variáveis mais importantes, como **Tags** e **Total Time Spent on Website**, para melhorar as conversões.
   - Ajustar campanhas de marketing ou comunicação com base nos insights obtidos.

2. **Diagnóstico do Modelo**:
   - Identificar se o modelo está tomando decisões com base em variáveis relevantes e interpretáveis.

## Qual é o valor de negócio deste modelo e porque devemos utilizá-lo?

Para atribuir valor de negócio a este modelo, vamos fazer algumas suposições:

- Um lead convertido em vendas gerará R$120 em receita no primeiro ano. (Valor arbitrário)
- O tempo e os esforços gastos na busca por leads priorizados (como previsto pelo modelo) custam R$15. (Valor arbitrário)
- Oportunidades perdidas pelo modelo (falsos negativos) geram um custo de oportunidade negativo de R$120 (você pode ou não incluir isso, pois não é um custo real, mas um custo de oportunidade – isso depende totalmente do caso de uso). (Valores arbitrários)

Se fizermos cálculos simples, chegaremos a um lucro de R$89.400. Este é o lucro da operação de conversão de leads mas qual o retorno específico de cada lead convertido? Neste caso hipotético a conversão é de R$120,00 mas e se for uma conversão de venda de equipamentos de energia solar no atacado? O lucro seria bem maior mas o custo da operação de conversão do lead seria o mesmo.

| **Descrição**                  | **Clientes**  | **Valor** | **Total**      |
|------------------------------  |---------------|-----------|-------------   |
| True Positive                  | 973           | $120      |  R$116.760,00   |
| True Positive + False Positive | 973+91=1064   | -$15      | -R$15.960,00       |
| False Negative                 | 95            | -$120     | -R$11.400,00       |
| Lucro total                    |               |           | **$89.400,00**    |


## Maximizar: Recall (Sensibilidade)
Recall mede a proporção de leads convertidos (positivos reais) que foram identificados pelo modelo. Isso ajuda a reduzir os falsos negativos (leads perdidos que poderiam ter sido convertidos). Se o custo de perder um lead relevante (falso negativo) é extremamente alto em relação ao custo de perseguir um lead irrelevante, priorize Recall.

## Maximizar: Precision (Precisão)
Precision mede a proporção de previsões positivas que realmente são positivas. Ou seja, ajuda a reduzir o número de falsos positivos (leads que não convertem, mas são perseguidos).
Se perseguir leads irrelevantes gera custos significativos que excedem o custo de oportunidade de perder leads, priorize Precision.


```python
modelo_tunado = exp.tune_model(melhor_modelo, optimize = 'Precision')
```

```python
modelo_producao = exp.finalize_model(modelo_tunado)
```

```python
# save model
exp.save_model(modelo_producao, 'modelo_producao_lead_conversion_2025_01_09')
```

```python
# load model
loaded_from_disk = exp.load_model('modelo_producao_lead_conversion_2025_01_09')
loaded_from_disk
```

## Uma vez o modelo treinado e salvo vamos efetuar o ranqueamento dos lead para disponibilizar para a equipe de vendas

```python
data_new = df.copy()
data_new.drop('Converted', axis=1, inplace=True)
data_new.head()
```

## Ranqueamento

### prediction_score_0 = probabilidade de **não** conversão

### prediction_score_1 = probabilidade de conversão

```python
resultados_ranqueamento = exp.predict_model(loaded_from_disk,data=data_new,raw_score=True)
```

```python
resultados_ranqueamento.head()
```

```python
resultados_ranqueamento['prediction_score_1'].max()
```

```python
resultados_ranqueamento['prediction_score_1'].min()
```

```python
resultados_ranqueamento = resultados_ranqueamento[['Lead Number','prediction_score_1','prediction_score_0','prediction_label']]
resultados_ranqueamento.sort_values('prediction_score_1', ascending=False, inplace=True)
resultados_ranqueamento.head()
```

```python
resultados_ranqueamento.to_csv('resultados_ranqueamento.csv', index=False)
```
